```python
import numpy as np
import matplotlib.pyplot as plt
import scipy, scipy.stats
```

```python
lower = scipy.stats.norm.ppf(0.01,1,1)
upper = scipy.stats.norm.ppf(0.99,1,1)
x = scipy.linspace(lower,upper,100)
```

```python
y = scipy.stats.norm.pdf(x,1,1)
```

```python
plt.plot(x,y)
```
