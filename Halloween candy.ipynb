## Halloween candy 

```python
import numpy as np
import pandas as pd
import scipy, scipy.stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("candy-data.csv") 
```

```python
data.head()
```

```python
sugar_win_df = data[["sugarpercent","winpercent"]]
price_win_df = data[["pricepercent","winpercent"]]
```

```python

sugar_win_df = sugar_win_df.sort_values(by='winpercent', ascending=True)
sugar_win_df.head()

x = sugar_win_df[["winpercent"]]
y = sugar_win_df[["sugarpercent"]]

price_win_df = price_win_df.sort_values(by='winpercent',ascending=True)

y_2 = price_win_df[['pricepercent']]


plt.figure(1,figsize=(10,8))
sugarplot= plt.subplot(211)
sugarplot.set_xlabel(xlabel="Win Percent")
sugarplot.set_ylabel(ylabel="Sugar percent")
plt.plot(x, y)  # sugar percent
priceplot = plt.subplot(212)
priceplot.set_ylabel(ylabel='Price percent')
plt.plot(x,y_2,color="green")# price percent




```

```python
choc_win_df = data[['chocolate','winpercent']]
choc_win_df = choc_win_df.loc[choc_win_df['chocolate'] == 1]
choc_win_df = choc_win_df.sort_values(by='winpercent',ascending=True)

choc_win_df.head()
```
