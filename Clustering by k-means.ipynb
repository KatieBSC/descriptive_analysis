```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python

x_values = np.random.normal(0,1,2500)
#x_2_values = np.random.exponential(scale=2,size=1000)
x_2_values = np.random.normal(5,.5,5000)
x_3_values = np.random.normal(10,30,2500)


#x_3_values = np.random.normal(10,12,1000)
#y_3_values = np.random.normal(10,12,1000)
```

```python
#Combine the two datasets
data_x = np.concatenate([x_values,x_3_values])
#data_x_2 = np.concatenate([x_2_values,y_2_values,y_2_values])
```

```python
#Create a DataFrame to hold the combined data from x's and y's
```

```python
dataset = pd.DataFrame()
```

```python
dataset['x'] = data_x
dataset['x_2'] = x_2_values

dataset.head(50)
```

```python
plt.figure(1,figsize=(10,10))
plt.scatter(dataset['x'] ,dataset['x_2'])
plt.show();
```

```python
kmeans = KMeans(n_clusters=4,random_state=0).fit(dataset)
dataset['cluster_nr'] = kmeans.predict(dataset)
```

```python
#Plot using scatter
plt.figure(1,figsize=(10,10))
plt.scatter(data_x,dataset['x_2'],c=dataset['cluster_nr'])
plt.show();
```
